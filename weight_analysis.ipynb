```python
import numpy as np
import pandas as pd
import openpyxl
from datetime import datetime
from openpyxl.utils.dataframe import dataframe_to_rows
```

```python
weight_data = pd.read_csv('./player_data/WA_weight_data.csv', index_col=['Name'])
```

```python
date_index = weight_data.columns[3:]
```

```python
WA = openpyxl.Workbook()
ws = WA.active
ws.title = 'WA_weight_team'
```

```python
for r in dataframe_to_rows(weight_data, index=True, header=True):
    ws.append(r)
```

```python
WA.save('./total_data/WA_weight.xlsx')
```

```python
for player in weight_data.index:
    pws = WA.create_sheet(player)
    weight = weight_data.loc[player][3:]
    w_df = pd.DataFrame(weight, index=date_index)
    for r in dataframe_to_rows(w_df, index=True, header=True):
        pws.append(r)  
```
