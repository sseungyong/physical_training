```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
now= datetime.now()
today = now.strftime('%Y-%m-%d')
print(today)
```

```python
train_data_01 = pd.read_csv('./player_data/KFA_MU-15_20190913_TR_AM.csv', index_col=['name'])
train_data_02 = pd.read_csv('./player_data/KFA_MU-15_20190914_TR_PM.csv', index_col=['name'])
```

```python
for player in train_data_01.index:
    player_path = './total_data/'+player+'.csv'
    player_data = train_data_01.loc[player:player].rename(index={player:today})
    player_data.to_csv(player_path, mode='a', header=False)
```
